```python
#Import dependency
import matplotlib.pyplot as plt
```

```python
#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Set the x-axis to a list of strings for each month.
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.plot(x_axis, y_axis)
#plt.xlabel('Months')
#plt.ylabel('Savings ($)')
#plt.title('Savings Account Results')
#plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, marker = "d", color = "green", linewidth = 2, label = "Boston")
ax.legend()
ax.set_xlabel('Months')
ax.set_ylabel('Fares ($)')
ax.grid()
ax.set_ylim(0, 45)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth =2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Add gridlines
plt.grid()
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color = "green", label = "Boston")
#label x & y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")
#Add a legend
plt.legend()
```

```python
#Matlab mathod
plt.barh(x_axis,y_axis, color = "magenta", label= "Boston")
plt.legend()
plt.xlabel('Date')
plt.ylabel('Fares ($)')
plt.gca().invert_yaxis()
```

```python
#Object oriented method, with ax.plt()
fib, ax=plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.set_xlabel("Months")
ax.set_ylabel("Fares ($)")
plt.gca().invert_yaxis()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c = "red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Months") 
plt.xlim(0, 45)
plt.gca().invert_yaxis()
```

```python
# Bubble Chart
#y_axis_larger = []
#for data in y_axis:
#    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])
```

```python
#Object oriented methor for Scatter/Bubble
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, s = [i*5 for i in y_axis], marker = 'o', c = '#87CEEB', alpha = 0.5, edgecolors = 'black', linewidth =2,  label="Boston" )
ax.set_ylabel("Months")
ax.set_xlabel("Fare ($)")
ax.set_title("PyBer Fare by Month") 
ax.set_xlim(0, 45)
ax.legend()
plt.gca().invert_yaxis()
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize = (8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode=explode_values, 
        colors = colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

```python
#Object oriented pie chart method
explode_values = (0, 0, 0.2, 0.1 , 0.1, 0.1 , 0.2, 0, 0, 0.1, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))

ax.pie(y_axis,
       labels=x_axis,
       explode = explode_values,
       colors = colors,
       shadow = True,
       startangle=270,
      autopct = '%.1f%%')
plt.gca().invert_yaxis()
plt.show()
```
