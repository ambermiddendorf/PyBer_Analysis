```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
file_to_read = "Resources/got.csv"

king_df = pd.read_csv(file_to_read)

king_df
```

```python
# Get attacker and defender king data
attackers = king_df.groupby(["attacker_king"]).count()["name"]
defenders = king_df.groupby(["defender_king"]).count()["name"]

attackers
```

```python
defenders
```

```python
kings_df = pd.DataFrame(
    {attackers: "Attackers",
    defenders: "Defenders"}, fillna(0))
kings_df
```

```python
# Get total battle data

battles_df = attackers.add(defenders, fill_value =0)
battles_df
```

```python
# Configure plot and ticks
battles_df.plot(kind = "bar", color ="red")
plt.xlabel("King")
plt.xticks(rotation=90)

# Set textual properties


# Show plot


# Resize plot to display labels
```
